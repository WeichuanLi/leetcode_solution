# 78. Subsets
Medium


Given a set of distinct integers, nums, return all possible subsets (the power set).

Note: The solution set must not contain duplicate subsets.

Example:

Input: nums = [1,2,3]
Output:
[
  [3],
  [1],
  [2],
  [1,2,3],
  [1,3],
  [2,3],
  [1,2],
  []
]

```python
class Solution(object):
    def dfs(self, nums, index,path):
       
        self.res.append(path)
        for i in xrange(index, len(nums)):
            self.dfs(nums, i+1, path+[nums[i]]) 
    def subsets(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        self.res = []
        self.dfs(nums, 0, [])
        return self.res
```

# 77. Combinations
Medium


Given two integers n and k, return all possible combinations of k numbers out of 1 ... n.

Example:

Input: n = 4, k = 2

Output:
[
  [2,4],
  [3,4],
  [2,3],
  [1,2],
  [1,3],
  [1,4],
]

```python
class Solution(object):
    def dfs(self,path,index,k,n):
        
        if len(path)==k:
            self.res.append(path)
           
        for i in xrange(index,n+1):
            
            
            self.dfs(path+[i],i+1,k,n)
          
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
    
        self.res=[]
        self.dfs([],1,k,n)
        return self.res
```

"123"

"213"
"231"
"312"
"321"
Given n and k, return the kth permutation sequence.

Note:

Given n will be between 1 and 9 inclusive.
Given k will be between 1 and n! inclusive.
Example 1:

Input: n = 3, k = 3
Output: "213"
Example 2:

Input: n = 4, k = 9
Output: "2314"

```python
class Solution(object):
    
    def getPermutation(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: str
        """
        
        f=[0]*n
        f[0]=1
        for i in xrange(1,n):
            
            f[i]=i*f[i-1]
          
        nums=[i for i in xrange(1,n+1)]
        
        res=""
        index=0
        k=k-1
        for i in range(n):
            
            index=k//f[n-1-i]
            k=k%f[n-1-i]
           
            res=res+str(nums[index])
            del nums[index]
            
        return res
```

  [5]
]


```python
class Solution(object):
    def dfs(self,target,path,candidates,start):
        
        if target<0:
            return 
        if target==0:
            self.res.append(path)
        for i in range(start,len(candidates)):
            if (i > start and candidates[i] == candidates[i - 1]):
                continue
            self.dfs(target-candidates[i],path+[candidates[i]],candidates,i+1)
    def combinationSum2(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.res=[]
        
        candidates.sort()

        self.dfs(target,[],candidates,0)
        return self.res
```

Example:

nums = [1, 2, 3]

target = 4

The possible combination ways are:

(1, 1, 1, 1)
(1, 1, 2)
(1, 2, 1)
(1, 3)
(2, 1, 1)
(2, 2)
(3, 1)

Note that different sequences are counted as different combinations.

Therefore the output is 7.

```python
class Solution(object):
    def dfs(self,nums,target,start,dic):
        
        if target<0:
            return 0
        if target==0:
            return 1
        if target in dic:
            return dic[target]
        res=0
        for i in xrange(len(nums)):
            res=res+self.dfs(nums,target-nums[i],i,dic)
        dic[target]=res
        return dic[target]
    def combinationSum4(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        self.res=0
        dic={}
        self.dfs(nums,target,0,dic)
        return dic[target]
```

```python
class Solution(object):
   
    def combinationSum4(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
       
        dp=[0]*(target+1)
        dp[0]=1
        
        for i in range(target+1):
            for j in nums:
                if i>=j:
                    dp[i]=dp[i]+dp[i-j]
        return dp[target]
```

Example 1:

Input: "aabb"

Output: ["abba", "baab"]

Example 2:

Input: "abc"

Output: []

```python
class Solution(object):
    
    def dfs(self,path,term,visited,final):
          
           if len(path)==len(term):
                self.res.append(path+final+path[::-1])
           for i in range(len(term)):
                if not visited[i]:
                    if i>0 and term[i]==term[i-1] and not visited[i-1]:
                        continue
                    visited[i]=True
                    self.dfs(path+term[i],term,visited,final)
                    visited[i]=False
        
    
    def generatePalindromes(self, s):
        """
        :type s: str
        :rtype: List[str]
        """
        if len(s)<=1:
            return s
        dic={}
        self.res=[]
        for i in s:
           
                dic[i]=dic.get(i,0)+1
        res=0
        term=""
        final=""
        for key,val in dic.items():
            
            if val%2!=0 :
                res=res+1
                final=key
                term=term+key*((val-1)/2)
            if val%2==0:
                term=term+key*(val/2)
           
        if res>1:
               
            return []
        visited=len(term)*[False]
        self.dfs("",term,visited,final)
        return self.res
```

# 980. Unique Paths III

On a 2-dimensional grid, there are 4 types of squares:

1 represents the starting square.  There is exactly one starting square.

2 represents the ending square.  There is exactly one ending square.

-1 represents obstacles that we cannot walk over.

Return the number of 4-directional walks from the starting square to the ending square, that walk over every non-obstacle square exactly once.

 

Example 1:

Output: 2

Explanation: We have the following two paths: 

1. (0,0),(0,1),(0,2),(0,3),(1,3),(1,2),(1,1),(1,0),(2,0),(2,1),(2,2)

2. (0,0),(1,0),(2,0),(2,1),(1,1),(0,1),(0,2),(0,3),(1,3),(1,2),(2,2)

Example 2:

Output: 4

Explanation: We have the following four paths:

2. (0,0),(0,1),(1,1),(1,0),(2,0),(2,1),(2,2),(1,2),(0,2),(0,3),(1,3),(2,3)

Example 3:

Input: [[0,1],[2,0]]

Output: 0

Explanation: 

There is no path that walks over every empty square exactly once.

Note that the starting and ending square can be anywhere in the grid.
 

Note:

1 <= grid.length * grid[0].length <= 20

```python
class Solution(object):
    def dfs(self,grid,x0,y0,non_empty):
        if not self.check(grid,x0,y0):
            return 
        if non_empty==0 and x0==self.end_1 and y0==self.end_2:
            self.res=self.res+1
            return 
        if self.check(grid,x0,y0):
            non_empty=non_empty-1
            grid[x0][y0]=-2
            self.dfs(grid,x0+1,y0,non_empty)
            self.dfs(grid,x0-1,y0,non_empty)
            self.dfs(grid,x0,y0-1,non_empty)
            self.dfs(grid,x0,y0+1,non_empty)
            
            non_empty=non_empty+1
            grid[x0][y0]=0
            
            
    def check(self,grid,x,y):
        return x>=0 and x< len(grid) and y>=0 and y<len(grid[0]) and grid[x][y]>=0
    def uniquePathsIII(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        start_1=0
        start_2=0
        self.end_1=0
        self.end_2=0
        non_empty=1
        self.res=0
        for i in range(len(grid)):
            
            for j in range(len(grid[0])):
                
                if grid[i][j]==0:
                    non_empty=non_empty+1
                if grid[i][j]==1:
                    start_1=i
                    start_2=j
                if grid[i][j]==2:
                    self.end_1=i
                    self.end_2=j
        self.dfs(grid,start_1,start_2,non_empty) 
        return self.res
```
